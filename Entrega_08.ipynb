## Prática 1: Ruído em Comunicações Analógicas

###  Sinal Real - Ruído real

```matlab
%% AWGN real
clc;clear all;close all;
% Parâmetros
load('Pratica_08_sinal_real.mat')
N=length(y);
t = 0:1/fs:(1/fs)*(N-1);               % Eixo do tempo          
x=Am*cos(2*pi*fm*t);               % Sinal original sem ruido x(t)
n=y-x;
%
% Geração manual das amostras de ruído 
Ps = sum(abs(x).^2)/N;                  % Calcula a potência do sinal
Pn = sum(abs(n).^2)/N;                  % Calcula a potência do ruído
SNR = Ps/Pn;                            % Calcula a SNR linear
SNR_dB = 10*log10(SNR);                 % Calcula a SNR em dB                       
%


% Mostrar informações
disp('Estimação de SNR: ')
disp(['   SNR do sinal real: ' num2str(SNR_dB) ' dB']);

%
% Gráficos
fig=figure;
subplot(3,1,1);
plot(t,x);
title('Sinal original')
axis([0 0.5 -4 4]);
%
subplot(3,1,2);
plot(t,y);
title('Sinal Com Ruido AWGN do arquivo');
axis([0 0.5 -4 4]);
%
subplot(3,1,3);
plot(t,n);
title('Ruido Retirado do Sinal');
axis([0 0.5 -2 2]);
%

fig.PaperUnits = 'inches';
fig.PaperPosition = [0 0 12 6];
```

## Prática 1: Ruído em Comunicações Analógicas

###  Sinal Complexo - Ruído Complexo

```matlab
%% AWGN real
clc;clear all;close all;
% Parâmetros
load('Pratica_08_sinal_complexo.mat')
N=length(y);
t = 0:1/fs:(1/fs)*(N-1);               % Eixo do tempo          
x=Ar*cos(2*pi*fm*t)+i*Ai*cos(2*pi*fm*t);                 % Sinal original sem ruido x(t)
n=y-x;
%
% Geração manual das amostras de ruído 
Ps = sum(abs(x).^2)/N;                  % Calcula a potência do sinal
Pn = sum(abs(n).^2)/N;                  % Calcula a potência do ruído
SNR = Ps/Pn;                            % Calcula a SNR linear
SNR_dB = 10*log10(SNR);                 % Calcula a SNR em dB                       
%


% Mostrar informações
disp('Estimação de SNR: ')
disp(['   SNR do sinal real: ' num2str(SNR_dB) ' dB']);

%
% Gráficos
fig=figure;
subplot(3,1,1);
plot(t,real(x));
title('Sinal original')
axis([0 0.5 -4 4]);
%
subplot(3,1,2);
plot(t,real(y));
title('Sinal Com Ruido AWGN do arquivo');
axis([0 0.5 -4 4]);
%
subplot(3,1,3);
plot(t,real(n));
title('Ruido Retirado do Sinal');
axis([0 0.5 -2 2]);
%

%
fig.PaperUnits = 'inches';
fig.PaperPosition = [0 0 12 6];
```

## Pratica 2: Ruídos em comunicações digitais

```matlab
% Parâmetros
EbN0_dB = 0;                                          % Eb/N0 de entrada
Ns = 10^5;                                             % Número de símbolos simulados

M = 2;                                                 % Número de símbolo da modulação
bits2 = randi([0 M-1],Ns,1);
txSig = pskmod(bits2,M);
rxSig = awgn(txSig,EbN0_dB);
% Gráficos
fig = figure;
subplot(1,3,1);
plot(rxSig(find(real(rxSig)>0)),'ro');
hold on;
plot(rxSig(find(real(rxSig)<=0)),'bs')
title('Diagrama de constelação pela função pskmod.m');
legend('Bit [0] transmitido','Bit [1] transmitido');
minAx = min([real(rxSig)' imag(rxSig)']);
maxAx = max([real(rxSig)' imag(rxSig)']);
axis([minAx maxAx minAx maxAx]);
% Sinal 8-PSK gerado manualmente

EbN0_dB = 5;  
M = 2;                                                 % Número de símbolo da modulação
bits2 = randi([0 M-1],Ns,1);
txSig = pskmod(bits2,M);
rxSig = awgn(txSig,EbN0_dB);
% Gráficos
subplot(1,3,2);
plot(rxSig(find(real(rxSig)>0)),'ro');
hold on;
plot(rxSig(find(real(rxSig)<=0)),'bs')
title('Diagrama de constelação pela função pskmod.m');
legend('Bit [0] transmitido','Bit [1] transmitido');
minAx = min([real(rxSig)' imag(rxSig)']);
maxAx = max([real(rxSig)' imag(rxSig)']);
axis([minAx maxAx minAx maxAx]);

% Sinal 8-PSK gerado pela função pskmod
EbN0_dB = 20;                                          % Eb/N0 de entrada
M = 2;                                                 % Número de símbolo da modulação
bits2 = randi([0 M-1],Ns,1);
txSig = pskmod(bits2,M);
rxSig = awgn(txSig,EbN0_dB);
% Gráficos
subplot(1,3,3);
plot(rxSig(find(real(rxSig)>0)),'ro');
hold on;
plot(rxSig(find(real(rxSig)<=0)),'bs')
title('Diagrama de constelação pela função pskmod.m');
legend('Bit [0] transmitido','Bit [1] transmitido');
minAx = min([real(rxSig)' imag(rxSig)']);
maxAx = max([real(rxSig)' imag(rxSig)']);
axis([minAx maxAx minAx maxAx]);
fig.PaperUnits = 'inches';
fig.PaperPosition = [0 0 24 12];
clear all;
```
